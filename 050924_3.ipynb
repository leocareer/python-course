```python
# Cargar un dataframe a partir de un fichero de excel, para leer un fichero de excel necesitaremos instalar una libreria,
# llamada openpyxl, un dataframe cargara solamente una hoja de excel (tenremos que especificar cual)
# openpyxl
```

```python
import pandas as pd
```

```python
ag2014 = pd.read_excel('VentasGroucery_Meses.xlsx', sheet_name='Ventas Agosto 2014')
ag2014
```

```python
sep2014 = pd.read_excel('VentasGroucery_Meses.xlsx', sheet_name='Ventas Septiembre 2014')
sep2014
```

```python
# Como anexar 2 dataframes para obtener uno solo, con todos los registros
# warning misma estructura de datos, mismo num. de colunas y con los mismos
# nombres, usaremos concat

bimensual = pd.concat([ag2014, sep2014], ignore_index=True)
bimensual
```

```python
bimensual['Amount'] = bimensual['Unit Price']*bimensual['Quantity']
bimensual
```
