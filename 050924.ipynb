```python
# El dataframe de pandas, sirve para trajajar con estructuras de 2 dimensione y se puedan cargar de 2 maneras,
# manualmente o leyendo de un fichero, base de datos u cualquier origen de datos

import pandas as pd   #pd as el acronimo

clase = pd.DataFrame([{'Nombre': 'Pedro', 'Estatura': 1.71, 'Edad': 28},{'Nombre': 'Maria', 'Estatura': 1.73, 'Edad': 45}, {'Nombre': 'Andrea', 'Estatura': 1.68, 'Edad': 28}, {'Nombre': 'Marcos', 'Estatura': 1.65, 'Edad': 19}])
print(clase)
```

```python
# Como ver solamente un campo o columna
clase['Nombre']
```

```python
clase['Edad'].sort_values(ascending=True)
```

```python
# Funciones de resumen
print(clase['Edad'].sum())
print(clase['Estatura'].max())
print('La persona de menos estatura',clase['Estatura'].min())
```

```python
# Promedio de edad, con redondeo
# Этот код вычисляет средний возраст и округляет результат до двух знаков после запятой.
print(round(clase['Edad'].mean(), 2))
```

```python
# Como anadir columnas a un Dataframe, tenemos 3 opciones
# 1) Poniendo un valor manual fijo
# 2) Haciendo un calculo a partir de una o varias columnas existentes
# 3) Aplicando una funcion co Apply
```

```python
# 1)
clase['Colegio']='IT Academy'
print(clase)
```

```python
# 2)
clase['Edad2'] = clase['Edad'] + 2
print(clase)
```

```python
# 3)
# Con aplly, restar 2 anos a los alumnos que tenga >30 anos
def resta2(edad):
  if edad > 30:
    return edad-2
  else:
    return edad

# Crear la columna nueva con apply
clase['Edad5'] = clase['Edad'].apply(resta2)
print(clase)
```

```python
clase['Diferencia'] = clase['Edad2'] - clase['Edad']
print(clase)
```

```python
# Incrementar la estatura de cada alumno en un 10%
clase['Estatura2'] = clase['Estatura'] + clase['Estatura']*0.1
print(clase)
```

```python
clase['Estatura3'] = clase['Estatura']*1.1
print(clase)
```

```python
# Como eleminar una columna, con la instruction drop + el parametro axis
# axis = 1, sianifica eliminar columnas
clase = clase.drop(['Estatura3'], axis = 1)
print(clase)
```

```python
clase.drop(['Edad2'], axis = 1, inplace=True)
print(clase)
```

```python
# Para eliminar una fila, lo mismo con drop, pero esta vez con axis = 0
clase.drop([2], axis=0, inplace=True)
print(clase)
```
